```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
data= pd.read_csv("houseprice.csv")
```

```python
data.head(3)
```

```python
plt.figure(figsize= (8, 8))
sns.heatmap(data= data.corr(), annot= True)
plt.show()
```

```python
x = data.iloc[:, 1:] 
y= data["price"]
```

```python
sc= StandardScaler()
sc.fit(x)
x= pd.DataFrame(sc.transform(x), columns= x.columns)
```

```python
x
```

```python
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.2, random_state= 42)
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
```

```python
# linear regression
```

```python
lr= LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
plt.figure(figsize= (15, 5))
plt.bar(x.columns, lr.coef_)
plt.show()
```

```python
# lasso for feature selection 
```

```python
la= Lasso(alpha = 0.5)
la.fit(x_train, y_train)
la.score(x_test, y_test)*100
```

```python
plt.figure(figsize= (15, 5))
plt.bar(x.columns, la.coef_)
plt.show()
```

```python
# ridge 
```

```python
ra= Ridge(alpha= 10)
ra.fit(x_train, y_train)
ra.score(x_test, y_test)*100
```

```python
plt.figure(figsize= (15, 5))
plt.bar(x.columns, ra.coef_)
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
```

```python
# linear regression(check error)
```

```python
mean_squared_error(y_test, lr.predict(x_test))
```

```python
mean_absolute_error(y_test, lr.predict(x_test))
```

```python
# root mean sq. error
np.sqrt(mean_squared_error(y_test, lr.predict(x_test)))
```

```python
# lasso
```

```python
print(mean_squared_error(y_test, la.predict(x_test)))
print(mean_absolute_error(y_test, la.predict(x_test)))
print(np.sqrt(mean_squared_error(y_test, la.predict(x_test))))
```

```python
# ridge
```

```python
print(mean_squared_error(y_test, ra.predict(x_test)))
print(mean_absolute_error(y_test, ra.predict(x_test)))
print(np.sqrt(mean_squared_error(y_test, ra.predict(x_test))))
```
