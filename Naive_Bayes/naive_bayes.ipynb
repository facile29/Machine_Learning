```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
data= pd.read_csv("placement_file.csv")
```

```python
data.head()
```

```python
# check its distribution, to get an idea which type of naive bayes model to apply
```

```python
plt.figure(figsize= (4, 3))
sns.kdeplot(data= data["cgpa"])
plt.show()
```

```python
plt.figure(figsize= (4, 3))
sns.kdeplot(data= data["iq"])
plt.show()
```

```python
plt.figure(figsize= (5, 3))
sns.scatterplot(x= "cgpa", y= "iq", data= data, hue= "placement")
plt.show()
```

```python
x= data.iloc[:, :-1]
y= data["placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.5, random_state= 42)
```

```python
# there are 3 types of naive bayes model
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
```

```python
gnb= GaussianNB() 
gnb.fit(x_train, y_train)
```

```python
gnb.score(x_test, y_test)*100
```

```python
gnb.score(x_train, y_train)*100
```

```python
mnb= MultinomialNB()
mnb.fit(x_train, y_train)
```

```python
mnb.score(x_test, y_test)*100, mnb.score(x_train, y_train)*100
```

```python
bnb= BernoulliNB()
bnb.fit(x_train, y_train)
```

```python
bnb.score(x_test, y_test)*100, bnb.score(x_train, y_train)*100
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf= gnb)
plt.show()
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf= mnb)
plt.show()
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf= bnb)
plt.show()
```

```python
# gaussian naive bayes gives us the good score and created a perfect decision boundary, perfect line 
```
