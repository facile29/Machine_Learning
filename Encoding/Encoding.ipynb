```python
# ordinal encoding(using skelarn)
```

```python
data= pd.DataFrame({'Size':['s','m','m','l','l','m','s','s']})
data
```

```python
order= [['s','m','l']]
```

```python
import sklearn 
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe= OrdinalEncoder(categories= order)
oe.fit(data[["Size"]])
```

```python
data["Size"]= oe.transform(data[["Size"]])
```

```python
data
```
